```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
PS4csv = "PS4_GamesSales.csv"
PS4_data_df = pd.read_csv(PS4csv)
```

```python
NewPS4_data_df = PS4_data_df[['Game', 'Year', 'Genre', 'Global']].copy()
NewPS4_data_df['Platform'] = "PS4"
RenamedPS4_data_df = NewPS4_data_df.rename(columns={'Game': 'game', 'Year':'year', 'Genre':'genre','Global':'ps4_global','Platform':'platform'})
```

```python
Xboxcsv = "XboxOne_GameSales.csv"
Xbox_data_df = pd.read_csv(Xboxcsv)
```

```python
NewXbox_data_df = Xbox_data_df[['Game', 'Year', 'Genre', 'Global']].copy()
NewXbox_data_df['Platform'] = "Xbox_One"
RenamedXbox_data_df = NewXbox_data_df.rename(columns={'Game': 'game', 'Year':'year', 'Genre':'genre','Global':'xbox_global', 'Platform':'platform'})
```

```python
RenamedPS4_data_df.dropna(how='any', inplace = True)
RenamedPS4_data_df
```

```python
RenamedXbox_data_df.dropna(how='any', inplace = True)
RenamedXbox_data_df
```

```python
rds_connection_string = "admin1:12345@localhost:5432/ETLProject"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
RenamedPS4_data_df.to_sql(name='ps4games', con=engine, if_exists='append', index=False)
```

```python
RenamedXbox_data_df.to_sql(name='xboxgames', con=engine, if_exists='append', index=False)
```
